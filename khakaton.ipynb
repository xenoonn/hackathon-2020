```python
def get_tab_count(dataset, x, y):
    return dataset[x & y].count()[0]
```

```python
from collections import Counter
import string
import nltk
import numpy as np
```

```python
import pandas as pd
import csv
disney = []
with open('princess_corpus.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
        disney.append(row)

del disney[0] 
```

```python
early_pr = []
mid_pr = []
late_pr = []
for line in disney:
    if line[2] == 'PRINCESS':
        text = line[1]
        char = line[3]
        if line[0] == 'EARLY':
            early_pr.append((text, char))
        elif line[0] == 'MID':
            mid_pr.append((text, char))
        elif line[0] == 'LATE':
            late_pr.append((text, char))
```

```python
phrase0 = [c[0] for c in early_pr]
phrase1 = [c[0] for c in mid_pr]
phrase2 = [c[0] for c in late_pr]
early_break = [nltk.tokenize.wordpunct_tokenize(text) for text in phrase0]
mid_break = [nltk.tokenize.wordpunct_tokenize(text) for text in phrase1]
late_break = [nltk.tokenize.wordpunct_tokenize(text) for text in phrase2]
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
#from nltk.stem.lancaster import LancasterStemmer
#stemmer = LancasterStemmer()
import spacy
# Initialize spacy 'en' model, keeping only tagger component needed for lemmatization
nlp = spacy.load('en', disable=['parser', 'ner'])
punctuation = string.punctuation # '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
# Add numbers
punctuation += '0123456789'

def clean(period):
    text = []
    for lista in early_break:
        nlp = spacy.load('en', disable=['parser', 'ner'])
        ll = ' '.join(lista)
        #print(ll)
        l = nlp(ll)
        lemmas = " ".join([token.lemma_ for token in l])
        for word in lemmas:
            rr = lemmas.split()
        
        valids = [word for word in rr if word != '-PRON-' and word not in stopwords and word not in punctuation 
                      and len(word)>2]


        valids_true = [''.join([char for char in str(word) if char not in punctuation]) for word 
                           in valids if len(''.join([char for char in str(word) if char not in punctuation]))>0]
        text.append(valids_true)
        return(text)
```

```python
early_clear = clean(early_break)
```

```python
mid_clear = clean(mid_break)
```

```python
late_clear = clean(late_break)
```

```python
def counter(clear):
    cnt = Counter()
    for words in clear:
        for word in words:
            cnt[word] += 1
    return cnt
```

```python
early_cnt = counter(early_clear)
mid_cnt = counter(mid_clear)
late_cnt = counter(late_clear)
```

```python
early_fin = early_cnt.most_common(10)
early_fin1 = early_cnt.most_common(30)
earlyy_fin = dict(early_fin)
```

```python
mid_fin = mid_cnt.most_common(10)
mid_fin1 = mid_cnt.most_common(30)
midd_fin = dict(mid_fin)
```

```python
late_fin = late_cnt.most_common(10)
late_fin1 = late_cnt.most_common(30)
latee_fin = dict(late_fin)
```
